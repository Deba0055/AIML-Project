# 2.A

```python
null_pct=df.isnull().sum()/df.shape[0]*100
for x in null_pct:
    if x>20:
        col_drop=null_pct[null_pct>20].keys()
        df1=df.drop(col_drop,axis=1)
```

```python
df1.head()
```

```python
df2=df1.fillna(df1.mean())
```

```python
df2.head()
```

# 2.B

```python
nunique=df2.nunique()
c_drop=nunique[nunique==1].index
df3=df2.drop(c_drop,axis=1)
```

```python
df3.head()
```

# 2.C

```python
df3['Time'].dtypes
```

```python
df4=df3.drop('Time',axis=1) # Not valid numbers for analysis
```

```python
df4.head()
```

# 2.D

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df4[list(df4.columns[:-1])]

vif_info = pd.DataFrame()
vif_info['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_info['Column'] = X.columns
vif_info
```

```python
for z in vif_info['VIF']:
    if z>10:
        df5 = df4.drop(vif_info['Column'].index,axis=0)
```

```python
df4.shape
```

```python
df5.shape
```

# 2.E

Observation and assumptions:

(2.B). We found out features which are having same value for all the rows and deleted the columns by using 'nunnique'.
       So reduction of columns from 560 to 444.
       
(2.C). Since column'Time' is not a number, we remove this column.

(2.D). We check for multi-collinearity by using VIF(Variance Inflation Factor). If VIF>10, then there is high 
        collinearity among the variables. Using for loop, these variables are deleted (variable with VIF>10).
        A significant reduction in number of rows from 1567 to 1125.
