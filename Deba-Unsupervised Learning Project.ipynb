### 5.A

Pre-requisite assumptions for PCA:

1. Linearity: Assumes the data set to be linear combinations of
the variables.

2. The importance of mean and covariance: There is no guarantee that the directions of maximum
variance will contain good features for discrimination.

3. That large variances have important dynamics: Assumes that components with larger variance
correspond to interesting dynamics and lower ones
correspond to noise.

### 5.B

Advantages of PCA:

1. We need to reduce the large number of features in our dataset. PCA does this efficiently.There is no correlation among principal components i.e, they are independent.

2. PCA gets rid of correlated variables which don’t contribute in any decision making. It improves the performance of algorithm.

3. It tackles overfitting by reducing the number of features.

Limitations of PCA:

1. Principal Components are not as readable and interpretable as original features.

2. Data standardization is neccessary before implementing PCA, otherwise PCA will not be able to find the optimal Principal Components.

3. PCA covers maximum variance. However, we tend to lose data if we are not careful while selecting the number of components of PCA.
