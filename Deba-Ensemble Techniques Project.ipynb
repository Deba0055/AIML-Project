```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#Function to plot Pie-Charts for all categorical variables in the dataframe
def pie_charts_for_CategoricalVar(cd_pie,m):
    '''Takes in a dataframe(cd_pie) and plots pie charts for all categorical columns. m = number of columns required in grid'''
    
    #get all the column names in the dataframe
    a = []
    for i in cd_pie:
        a.append(i)
    
    #isolate the categorical variable names from a to b
    b = []
    for i in a:
        if (cd1[i].dtypes.name) == 'category':
            b.append(i)
        
    plt.figure(figsize=(15, 12))
    plt.subplots_adjust(hspace=0.2)
    plt.suptitle("Pie-Charts for Categorical Variables in the dataframe", fontsize=18, y=0.95)
    
    # number of columns, as inputted while calling the function
    ncols = m
    # calculate number of rows
    nrows = len(b) // ncols + (len(b) % ncols > 0)
    
    # loop through the length of 'b' and keep track of index
    for n, i in enumerate(b):
        # add a new subplot iteratively using nrows and ncols
        ax = plt.subplots(nrows, ncols, n + 1)

        # filter df and plot 'i' on the new subplot axis
        cd1.groupby(i).size().plot(kind='pie', autopct='%.2f%%',ax=ax)
        
        ax.set_title(i.upper())
        ax.set_xlabel("")
        ax.set_ylabel("")
    plt.show()
```

```python
def preprocess_data(cd,Churn):
    #Filling the null values
    cd1 = cd
    cd1['TotalCharges']=cd1['TotalCharges'].fillna(cd1['TotalCharges'].mean())
    #Converting all continuous values to float
    cd1['TotalCharges'] = pd.to_numeric(cd1['TotalCharges'], errors='coerce',downcast='float')
    cd1['tenure'] = pd.to_numeric(cd1['tenure'], errors='coerce',downcast='float')
    cd1['SeniorCitizen'] = pd.to_numeric(cd1['SeniorCitizen'], errors='coerce',downcast='float')
    cd1['MonthlyCharges'] = pd.to_numeric(cd1['MonthlyCharges'], errors='coerce',downcast='float')
    #Dropping unnecessary feature
    cd_new=cd1.drop(['customerID'],axis=1)
    #Encoding categorical values
    replaceStruct = {
                'OnlineBackup':{'Yes': 1, 'No': 0,'No internet service': 2},
                'DeviceProtection':{'Yes': 1, 'No': 0, 'No internet service': 2},
                 'TechSupport':{'Yes': 1,'No':0 ,'No internet service': 2},
                 'StreamingTV':{'Yes': 1, 'No': 0 ,'No internet service': 2},
               'StreamingMovies':{'Yes': 1, 'No': 0,'No internet service':2},
               'MultipleLines':{'Yes': 1, 'No': 0,'No phone service':2},
                'InternetService':{'DSL': 1,'Fiber optic': 2,'No':0},
                  'Contract':{'Month-to-month':0,'One year':1,'Two year':2},
                    'PaperlessBilling':{'Yes':1,'No':0},
                    'PaymentMethod':{'Electronic check':0,'Mailed check':1,'Bank transfer (automatic)':2,
                                     'Credit card (automatic)':3},
                   'gender':{'Male':1,'Female':0},
                   'Partner':{'Yes':1,'No':0},
                   'Dependents':{'Yes':1,'No':0},
                   'PhoneService':{'Yes':1,'No':0},
                   'OnlineSecurity':{'Yes':1,'No':0,'No internet service':2},
                    'Churn':{'Yes':1,'No':0}}
    oneHotCols=['SeniorCitizen']
    cd_new=cd_new.replace(replaceStruct)
    cd_new=pd.get_dummies(cd_new, columns=oneHotCols)
    return cd,Churn
```
